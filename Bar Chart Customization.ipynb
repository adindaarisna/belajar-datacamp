```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
plt.bar(xdat, ydat)
plt.title('Vertical Bars')
plt.show()
```

```python
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
plt.barh(xdat, ydat)
plt.title('Horizonal Bars')
plt.show()
```

```python
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
barcolor=['cyan','green','grey','blue','yellow','lightgreen']
plt.barh(xdat, ydat, color=barcolor, edgecolor='red')
plt.title('Colored Bars')
plt.show()
```

```python
# Antar bar ada jarak
# tapi belum ada label di sumbu x nya
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
pos=[0,1,5,8,9,10]
plt.bar(pos, ydat)
plt.title('Seperated Bars')
plt.show()
```

```python
# versi sudah ada label di sumbu x
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
posisi=[0,1,5,8,9,10]
plt.bar(posisi, ydat)
# Create names on the x-axis
plt.xticks(posisi, xdat)
plt.title('Seperated Bars')
plt.show()
```

```python
# Custom Labels
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
plt.bar(xdat, ydat)
plt.title('Vertical Bars')
#Create name on the x-axis
plt.xticks(color='red', fontsize=15)
plt.yticks(color='blue', fontsize=10)
plt.show()
```

```python
# Custom Margins
ydat=[5,9,20,16,30,3]
xdat=['Adinda','Budi','Chloe','Dashie','Ellena','Frank']
plt.bar(xdat, ydat)
plt.title('Vertical Bars')
#Create name on the x-axis
plt.xticks(rotation=-35)
plt.yticks(color='blue', fontsize=10)
# Custom dimensi tabelnya. Bisa dibetot jadi lebih tinggi wkwk
plt.subplots_adjust(bottom=1, top=2)
plt.show()
```

```python
# Add Number/Words in every bar
ydat=[5,9,20,16,30,3]
xdat=['A','B','C','D','E','F']
xposition=[0,1,2,3,4,5]
plt.bar(xdat, ydat)
plt.title('Vertical Bars')
plt.text(xposition[0]-0.3, ydat[0]+0.5, s='Adinda', size=10)
plt.text(xposition[1]-0.3, ydat[1]+0.5, s='Ben', size=10)
plt.text(xposition[2]-0.3, ydat[2]+0.5, s='Chole', size=10)
plt.show()
```

```python
# Stack Bar Chart
medals=pd.read_csv('medals_by_country_2016.csv', index_col=0)
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"], label="Gold")
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"],
label="Silver")
ax.bar(medals.index, medals["Bronze"],
bottom=medals["Gold"] + medals["Silver"],
label="Bronze")

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Number of medals")
ax.legend()
plt.show()
```
